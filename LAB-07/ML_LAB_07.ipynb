```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

```

```python
df = pd.read_csv('6 class.csv')
df.head()
```

```python
print(df.columns)
print(df.shape)
print(df.isnull().sum())
```

```python

df = pd.get_dummies(df, columns=['Star color', 'Spectral Class'], prefix=['Star_color', 'Spectral_Class'])
```

```python
X = df.drop('Star type',axis=1)
y = df['Star type']

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size = 0.3, random_state = 1)
```

```python
model = SVC(kernel='linear')
model.fit(Xtrain,ytrain)
predict = model.predict(Xtest)
report = classification_report(ytest,predict)
print(report)
```

```python
print("Number of support vectors:", len(model.support_))
```
